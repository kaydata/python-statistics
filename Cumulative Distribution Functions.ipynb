# Cumulative Distribution Functions (CDFs)

## Introduction to CDF

A Cumulative Distribution Function (CDF) is a probability function that provides the cumulative probability of a random variable being less than or equal to a given value. It is a fundamental concept in probability theory and statistics.

### Calculating CDF

Let's start by calculating CDF manually

```python
import numpy as np

# Sample dataset
data = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 10]

# Calculate CDF manually
sorted_data = np.sort(data)
cdf = np.cumsum(sorted_data) / np.sum(sorted_data)

# Print the CDF values
print("Sorted Data:", sorted_data)
print("CDF:", cdf)
```

This line displays the calculated CDF values corresponding to each data point in the sorted dataset. Each value represents the cumulative probability of a data point being less than or equal to its value.

For example:

- The first element in the sorted dataset is 1, and its CDF value is 0.01666667. This means that approximately 1.67% of the data points in the dataset are less than or equal to 1.
- The fifth element in the sorted dataset is 5, and its CDF value is 0.25. This indicates that approximately 25% of the data points in the dataset are less than or equal to 5.
- The last element in the sorted dataset is 10, and its CDF value is 1. This implies that 100% of the data points in the dataset are less than or equal to 10, as it is the maximum value in the dataset.

### CDF in Python

```python
import numpy as np
from scipy.stats import norm

# Generate a normal distribution
mu = 0
sigma = 1
data = np.random.normal(mu, sigma, 1000)

# Calculate CDF using built-in functions
cdf = norm.cdf(data, mu, sigma)

# Print the CDF values
print("CDF:", cdf)
```

### Visualizing CDF

```python

import matplotlib.pyplot as plt

# Generate a random dataset
data = np.random.normal(0, 1, 1000)

# Sort the data
sorted_data = np.sort(data)

# Calculate the CDF
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Plot the CDF
plt.plot(sorted_data, cdf)
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Data")
plt.ylabel("Cumulative Probability")
plt.show()
```

### Comparing CDFs


```python

# Generate two datasets
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(1, 1.5, 1000)

# Sort the data
sorted_data1 = np.sort(data1)
sorted_data2 = np.sort(data2)

# Calculate the CDFs
cdf1 = np.arange(1, len(sorted_data1) + 1) / len(sorted_data1)
cdf2 = np.arange(1, len(sorted_data2) + 1) / len(sorted_data2)

# Plot the CDFs
plt.plot(sorted_data1, cdf1, label="Dataset 1")
plt.plot(sorted_data2, cdf2, label="Dataset 2")
plt.title("Comparing CDFs")
plt.xlabel("Data")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.show()
```
